```python
plt.figure(figsize = (10,6))
sns.histplot(data=df, x="QI 2024" , kde = 'dist')
```

```python
outliers= ['QI 2024']
plt.rcParams['figure.figsize']=[4,4]
sns.boxplot(data=df[outliers],orient ="v",palette="husl", width= 0.7)
plt.title("outliers")
plt.ylabel(" QI 2024 Range")
plt.show
```

```python
round((df['Revenues'].value_counts()/df.shape[0])*100,1)
```

```python
colors = sns.color_palette("ch:start=.2,rot=-.3")[0:5]
plt.figure (figsize=(10,7))
plt.pie(round((df['Revenues'].value_counts()/df.shape[0])*100,1), labels = df['Revenues'].unique(), colors = colors, autopct='%.0f%%')
plt.show()
```
